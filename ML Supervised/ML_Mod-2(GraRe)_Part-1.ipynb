### What is Gradient Descent in Linear Regression?
- It is a optimization algorithm used in linear regression to find the best-fit line for the data.
- It works gradually adjusting the line's slope and intercept to reduce the difference between actual and predicted values.
- This process helps the model to accurate predictions by minimizing the errors step by step.

### Why Use Gradient Descent for Linear Regression
- For linear regression, we use cost function usually Mean Squared Error(MSE).To find the model parameter.
- For simple linear regression, we use normal Equation find parameters directly.For large datasets (or) high-dimensional data these methods become computationally expensive due to large matrix compuations and memory limitations.
- For ploynomial regressions, the cost function becomes complex and non-linear, so analytical solutions are not available.That's where gradient descent plays an important role even for large datasets and complex, high-dimensional problems. 

### How Does Gradient Descent Work in Linear Regression?
#### Steps involves :
#### 1. Initializing parameters:
- Start with random initial values for the slope(m) and intercept(b).
#### 2. Calculate the cost function:
- Measure the error by ising Mean Square Error(MSE).
- Formula - J(m, b) = 1/n Σ(y(i)-(mx(i)+b))²
#### 3. Compute the gradient :
- Calculate how much cost function changes with respect to m and b.
- For slope m - ∂J/∂m = 2/n Σx(i)(y(i)-(mx(i)+b))
- For intercept b - ∂J/∂b = 2/n Σ(y(i)-(mx(i)+b))
#### 4. Update Parameters:
- Change m and b to reduce the error.
- For slope m - m = m-α.∂J/∂m.
- For intercept b - b = b-α.∂J/∂b.
- Here α is the learning rate that controls the size of each updates.
#### 5. Repeat:
- Keep repeating steps 2-4 until the error stops decresing significantly.

### Implementation of Gradient Descent in Linear Regression
#### X, y = make_regression(n_samples=100, n_features=1, noise=15, random_state=42):
- Generating 100 data points with one feature and some noise for realism.
- random_state=42 - It is guarantee that the same datapoints will be included in the training and testing sets every time you run the code.
#### X_b = np.c[np.ones((m,1)),X]:
- Adding a column of ones to X to account for the intercept term in the model.
#### theta = np.array([[2.0],[3.0]]):
- Initializing model parameters ( intercept and slope) with starting values.


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression

X, y = make_regression(n_samples=100, n_features=1, noise=15, random_state=42)
y = y.reshape(-1,1)
m = X.shape[0]

X_b = np.c_[np.ones((m,1)), X]

theta = np.array([[2.0],[3.0]])

plt.figure(figsize=(10,5))
plt.scatter(X, y, color='blue', label='Actual data')
plt.plot(X, X_b.dot(theta), color="green", label="Initial Line (No GD)")
plt.xlabel("Feature")
plt.ylabel("Target")
plt.title("Linear Regression Without Gradient Descent")
plt.legend()
plt.show()
```

- Here the model’s predictions are not accurate and the line does not fit the data well. This happens because the initial parameters are not optimized which prevents the model from finding the best-fit line.

#### Now we will apply gradient descent to improve the model and optimize these parameters
#### learning_rate = 0.1, n_iterations = 100 :
- Set the learning rate and number of iterations for gradient descent to run respectively.
#### gradient = (2/m) * X_b,T.dot(y_pred - y):
- Finding gradients of the cost function with respect to parameters.
#### theta = theta - learning_rate * gradients :
Updating parameters by moving opposite to the gradient direction.

```python
learning_rate = 0.1
n_iterations = 100

for _ in range(n_iterations):

    y_pred = X_b.dot(theta)

    gradients = (2 / m) * X_b.T.dot(y_pred - y)

    theta = theta - learning_rate * gradients

plt.figure(figsize=(10, 5))
plt.scatter(X, y, color="blue", label="Actual Data")
plt.plot(X, X_b.dot(theta), color="red", label="Optimized Line (With GD)")
plt.xlabel("Feature")
plt.ylabel("Target")
plt.title("Linear Regression With Gradient Descent")
plt.legend()
plt.show()
```
