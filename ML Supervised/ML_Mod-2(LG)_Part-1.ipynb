### What is supervised Learning ?
- Trains models on labelled data to predict (or) claasify new, unseen data.
- Every input has a corresponding correct output.To make accurate predictions on new, unseen data.
- The model makes predictions and compares them with the true outputs, adjusting itself to reduce errors and improve accuracy overtime.
### Two Types of supervised Learning:
#### 1. Regression:
- Where the aim is to predict continous numerical values.
- Eg - Predicting a student's exam score based on the number of hours they studied.
#### 2. Classification:
- Where the goal is to predict discrete (or) categorical values.
- Eg - E-mail spam filtering, where the model predicts wheather an email is spam (or) not spam based on features like keywords, sender details, and message content.


### Most commonly used supervised learning algorithms are :
#### 1. Linear Regression:
- It is a supervised machine learning.
- It learns from the labelled datasets and maps the data points with most optimized linear functions which can be used for prediction on new datasets.
- It is represnted by straightline.
- It assumes that there is a linear relationship between the input and output, meaning the output changes at a constant rate as the input changes.
#### Real-world Use cases in linear regression:
- stock Market Prediction
- Real Estate Price Prediction
- Medical Risk Prediction
- Sales Forecasting.
#### Two Types of Linear Redression:
#### 1. Simple Linear Regression (or) Univariate regression:
- It is used when we want to predict a target value(dependent variable) using only one input feature (independent variable).
- Formula - Y = mX + C.
  - Here Y - Predicted value.
  - m - slope (or) coefficient.
  - C - intercept (  value of Y when X=0)
- Eg - Predicting person's salary(y) based on their years of experience(x).
#### 2. Multilinear Regression (or) Bivariate regression:
- It involves more than one independent variables and one dependent variable.
- The Equation will Be - Y = a1x1 + a2x2 + a3x3 + ...... +anxn + b.
     -  Y - Predicted value.
     - a1, a2,a3,... an - Coefficients (weights) corresponding to each predictor.
     - x1,x2,x3 .....xn - independent variables.
     - b - intercept.
- Eg - Predicting house prices.

### Assumptions in Linear regession:
-  1. Linearity - The relationship between inputs(x) and the output(Y) is a straight line.
   2. Independence of errors - The errors in predictions should not effect each other.
   3. Constant Varience(Homoscedasticity) - The errors should have equal spread across all values of the input. If the spread changes (like fans out (or) shrinks), it's called heteroscedasticity and it's a problem for the model.
   4. Normality of Errors - The errors should follow a normal (bell-shaped) distribution.
   5. No Multicollinearity (for multiple regression) - Input variables shouldn't be too closely related to each other.
   6. No Autocorrelation - Errors shouldn't show repeating patterns, especially in time-based data.
   7. Additivity - The Total effect Y is just the sum of effects from each X, no mixing (or) interaction between them.

### Best fit line in Linear regression:
- It is the straight line that most accurately represents the relationship between independent variable(input) and the dependent variable(output).
- It is the line that minimizes the difference between the actual data points and the predicted values from the model.
#### Minimizing the Error: The Least Squares Method 
- To find best line, we use a method called Least squares.
- This method is to minimize the sum of sqaured difference between the actual values(data points) and predicted values from the line.
- Formula - Sum of squarederrors(SSE) = Σ(Y(i) - Ȳ(i))²
- Where, - Y(i) - Actual observed value, Ȳ(i) - Predicted value .
#### Interpretation of the Best-Fit line:
#### Slope(m):
- The slope of the best-fit line indicates how much the dependent variable(y) chnages with each unit change in the independent variable(y).
- Eg - if the slope is 5, it means that for every 1-unit increase in x, the value of y increases by 5 units.
#### Intercept(b):
- It represents the predicted value of y when x=0. It's the point where the line crosses the y- axis.

### Cost function for Linear regression
- In LinRe, It measures how far the predicted values are from the actual values(Y).
- It helps identify and reduce errors to find the best-fit line.
- The most common cost function used are Mean Squared Error(MSE) and Mean Absolute Error(MAE).
#### 1. Mean Squared Error (MSE):
- It measures the average squared difference between the predicted value and actual value.
- Formula - MSE = 1/n Σ(Y(i) - Ȳ(i))²
    - n - number of data points.
    - Y(i) - actual (or) observed value for the ith data point.
    - Ȳ(i) - predicted value for the ith data point.

#### 2. Mean Absolute Error (MSE):
- It measures the average absolute difference between the predicted value and actual value.
- It is less sensitive to outliers than MSE.
- Formula - MSE = 1/n Σ(Y(i) - Ȳ(i))²
    - n - number of data points.
    - Y(i) - actual (or) observed value for the ith data point.
    - Ȳ(i) - predicted value for the ith data point.
#### 3. Root Mean Square Error(RMSE):
- The square root of the residual's varience is the Root Mean Square Error.
- It describes how well observed data points match the expected values (or) the model's absolute fit to the data.
- It can be expressed as - RMSE = √(RSS /n) = √1/nΣ(P(i) -O(i))²
- Where,
    -  P(i) - Predicted value.
    -  O(i) - actual value.
    -  n - no of observation points.
#### 4. Coefficient of Determination(R-squared):
- It is a statistical measure used in machine learning to evaluate the performance of regression models.
- It is always in the range of 0 to 1.
- In general, the better the model matches the data, the greater the R-squared number.
- Formula - R² = 1 - (RSS/TSS)
     - RSS - Residual sum of squares.
     - TSS - Total sum of squares.
#### a.Residual sum of squares:
- The sum of squared difference between the actual values and predicted values of the response variable.
- It quantifies prediction error.
- Formula - RSS = Σ(Y(i) - Ȳ(i))²
#### b.Total sum of squares:
- It measures total variation.
- The sum of squared difference between the observed values of the response variable for observation i, and Ȳ is the mean of the observed values.
- Formula - TSS = Σ(Y(i) - Ȳ)².
#### 5. Adjusted R-Squared Error:
- It ranges from 0.5 to 1.0
- Before finding this, we need to find R(squared) value.
- It measures the proportion of variance in the dependent variable that is explained by independent variables in a regression model.
- Formula - Adjusted R² = 1 - [(1-R²)(n-1)/n-p-1)]
    - R² - r-squared of the model
    - n - number of observation(data points)
    - p - number of predictors (independent variables) 

### Python Implementation of Linear Regression
#### 1. Import the necessary libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

#### 2. Generating Random Dataset
- Fetches the calfornia Housing datset and separates features(X) and target(y).
- np.randon.seed() - function sets the seed for NumPy's pseudo-random number generator.
- np.random.rand() - Produces values from a uniform distribution over half-open interval[0.0,0.1).
- np.random.randn() - produces samples from the Normal Distribution with mean of 0 and variance of 1.

```python
np.random.seed(42)
X = np.random.rand(50, 1) * 100  
y = 3.5 * X + np.random.randn(50, 1) * 20 
```

#### 3. Creating and Training Linear Regression Model


```python
model = LinearRegression()
model.fit(X, y)
```

#### 4. Predicting Y Values

```python
Y_Pred = model.predict(X)
```

#### 5. Visualizing the Regression Line

```python
plt.figure(figsize=(8,6))
plt.scatter(X, y, color='blue', label='Data Points')
plt.plot(X, Y_Pred, color='red', linewidth=2, label='Regression Line')
plt.title('Linear Regression on Random Dataset')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
