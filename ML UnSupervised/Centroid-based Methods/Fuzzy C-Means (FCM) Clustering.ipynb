### Fuzzy C-Means Clusering:
- Fuzzy clustering allows each data point to belong multiple clusters with different membersjip values.
- Instead of assigning a point to just one group, it captures how stringly a point relates to each cluster.
  1. Uses membership scores between 0 and 1.
  2. Handles overlapping or unclear cluster boundaries.
  3. More flexible than hard clustering methods.
  4. Useful when data don't fit neatly into a single group.

### Working of Fuzzy Clustering:
#### Step 1: Initialize Membership Values Randomly:
- Each data point is given random membership scores for all clusters. A point can partially belong to multiple clusters.
#### Step 2: Compute cluster centroids:
- Centroids are calculated as weighted averages, which weights are membership values raised to the fuzziness parameter m.
#### Step 3: Calculate Distance between data points and centroids:
- Calculate the Euclidean distance between each point and every centroid to dtermine proximity, which will be used for to update memberships.
#### Step 4: Update Membership Values:
- Membership values are updated inversly proportional to the these distances. Points closer to a centroid get higher membership.
#### Step 5: Repeat Until Convergence:
- Steps 2-4 are repeated until the membership values stabilize meaning there is no significant changes from one iteration to another.
- This indicate that the clustering has reached an optimal state.
