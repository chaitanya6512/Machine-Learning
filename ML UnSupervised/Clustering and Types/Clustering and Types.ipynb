### What is Unsupervised Learning?
- Finds patterns or groups in unlabelled data like clustering or dimensionality reduction.
- It can be again divided into three categories based on their purpose:
  1. Clstering
  2. Dimensionality reduction
  3. Association Rule Mining

### What is Clustering?
- It is an Unsupervised machine Learning technique that groups similar data points together into cluster based on their characteristics, without using any labelled data.
 1. Goal -Discover the natural grouping or structure in unlabelled data without predefined categories.
 2. How - Data points are assigned to clusters based on similarity or distance measures.
 3. Similarity Measure - Can include Euclidean distance, cosine similarity or other metrics depending on data type and clustering method.
 4. output - Each group is assigned a cluster ID, representing shared characteristics within the cluster.
### Types of Clustering:
#### 1. Hard Clustering:
- In hard clustering, each data point strictly belongs to exactly one cluster, no overlap is allowed.
- This approach assigns a clear membership, making it easier to interpret and use for definitive segmentation tasks.
- Example - If clustering customer data into 2 segments. each customer belongs fully either Cluster1 or Clster 2 without partial memberships.
- Use cases - Market segmentation, customer grouping, document clustering.
- Limitations - Cannot represent ambiguity or overlap between groups.
#### 2. Soft Clustering:
- Soft clusterings assigns each data point a probability or degree of membership to multiple clusters simutaneously, allowing data points to partially belong to several groups.
- Example - A data point may have 70% membership in cluster 1 and 30% membership in cluster 2, reflecting uncertainity or overlap in group characteristics.
- Use cases - Situations with overlappig class boundaries, fuzzy categories like customer personas or medical diagnosis.
- Benefits - Captures ambiguity in data, models gradual transitions between clusters.

### Types of Clustering Methods:
#### 1. Centroid-based Clustering(Partioning Methods):
- Centroid-based Clustering organizes data as central protypes called centroids, where each cluster is represented by the mean(or medoid) of its members.
- The number of cluster is specified in advance and the algorithm allocates data points into nearest clusters, making this technique efficient for spherical and similarly sized clusters but sensitive to outliers and initialization.
- Algorithms -
  1. K-means - Iteratively assigns points to nearest centroid and recalculates centroid to minimize intra-cluster variance.
  2. K-medoids - Similar to K-means cluser but uses as actual data points (medoids) as centers, robust to outliers.
#### 2. Density-based Clustering(Model-based Methods):
- Density-based clustering defines clusters as contiguous regions of high data density seperated by areas of lower density.
- This approach can identify clusters of arbitrary shapes, handles noise well and does not require predefining the number of clusters, through its effectiveness depends on chosen density parameters.
- Algorithms -
  1. DBSCAN [ Density-Based Spatial clustering of applications with Noise ] - Groups points with sufficent neighbors. labels sparse points as noise.
  2. OPTICS [  Ordering Points To Identify Clustering Structure ] - Extends DBSCAN to handle varying densities.
#### 3. Connectivity-based Clustering (Hierarchical clustering):
- Connectivity-based clustering builds nested grouping of data by evaluating how data points are connted to their neighbors.
- It creates a dendogram a tree like structure that reflects relationships at various granularity levels and does not require specifying cluster number in advance, but can be computationally intensive.
- Algorithms -
  1. Aggolmerative(Bottom-up) - Start with each point as a cluster; iteratively merge closest clusters.
  2. Divisive(Top-down) - Start with one cluster; iteratively split into smaller clusters.
#### 4. Distributed-based Clustering:
- Distributed-based clustering assumes data is generated from a mixture of probability distributions such as Gaussian distribution.
- It assigns points to clusters based on statistical likelihood.
- This method support clusters with flei=xible shapes and overlaps, but usually requires specifying the number of distributions.
- Algorithm -
  1. Gaussian Mixture Model(GMM) - Fits data as weighted mixture of Gaussian distributions; assigns data based on likelihood.
#### 5. Fuzzy Clustering:
- Fuzzy clustering extends traditional methods by allowing each data point to belong to multiple clusters with varying degrees of membership.
- This approach captures ambiguity and soft boundaries in data.
- It is particularly useful when the clusters overlap or boundaries are not clear-cut.
- Algorithms -
  1. Fuzzy C-Means - Similar to K-means but with fuzzy memberships updated iteratively.

### Use Cases:
1. Customer segmentation - Grouping customers based on behaviour or demographics for targeted marketing and personalized services.
2. Anomaly Detection - Identiying outliers or fraudulent activities in finance, network security and sensor data.
3. Image Segmentation - Dividing images into meaningful parts for object detection, medical diagnostics or computer vision tasks.
4. Recommendation System - Clustering user preference to recommend movies, products or content tailored to different groups.
5. Market Basket Analysis - Discovering products frequently bought together to optimize store layouts and promotions.
