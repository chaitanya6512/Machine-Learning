### Feature Extraction
- It is a Process of transforming raw data  into a simplified and informative set of features (or) attributes.
- Improving model accuracy and reducing computational costs by  focusing on essential aspects of the data.
- This reduce cpmplexity and highlights the most relevant information making it easier for machine learning models to analyze and learn from the data efficiently.

#### Key Techniques of Feature Extraction:
#### 1. Statistical Methods:
- To summarixe and explains patterns of data.
- Mean - The average of data values in the dataset.
- Median - The Mid value when it is sorted in ascending order.
- Standard Deviation - A measure of the spread (or) dispersion of sample.
- Correlation and Covarience - Measures of the linear relationship between two (or) more factors.
- Regression Analysis - A way to model the link between a dependent variable and one-more independent variables.
#### 2. Dimensionality Reduction
- It reduces the number of features without losing important features.
#### Principal Component Analysis (PCA) 
- It is an unsupervised machine learning technique used for dimensionality reduction in datasets with a large number of features.
- While keeping the most important features.
#### Linear Discriminant Analysis
- It ia also called Normal Discrminant analysis.
- It is a supervised classification problem that helps seperate one (or) more classes by converting higher-dimensional data to lower-dimensional data.
- It is used for linear combination of features that best separates classes with in a dataset.
#### T-distributed Stochastic Neighbor Embedding (t-SNE) Algorithm
- It is a non linear dimensionality reduction technique used for visualizing high-dimensional in a small-dimensional space on mainly in 2D (or) 3D.
- In other words, A technique that reduces high-dimensional data into two (or) more dimensions ideal for visualizing complex datasets.

#### 3. Feature Extraction for Textual Data
- In Natural language processing(NLP), we often convert raw text into a format that machine learning models can understand.
#### Bag of Words(BoW)
- Represents a document by counting word frequencies, ignoring word order, useful for basic text classification.
#### Term Frequency-Inverse Document frequency(TF-IDF)
- Adjusts word importance based on a frequency in a specified document compared to all documents, highlights unqiue terms.

#### 4. Signal Processing Methods
- It is used for analyzing audio, time series and sensor data.
#### Fourier Transform 
- It converts signal from time domain to the frequency domain to analyze its frequency components.
#### Wavelet Transform
- It analyzes signals that vary over time, offering both time and frequency information for non-stationary signals.

#### 5. Image Data Extraction
- Techniques used for extracting features from images :
#### Histogram of oriented Gradients(HoG)
- It is used in object detection (or) recognition tasks.
- This technique finds distribution density (or) edge direction in an image.
#### Convolutional neutral networks(CNN)
- They learn hierarchical features from images through layers of conventionals, ideal for classification and detection tasks.
